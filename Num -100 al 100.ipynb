```
def frange(start, stop, step):
    start = float(start)
    count = 0
    while True:
        temp = float(start + count * step)
        if step > 0 and temp >= stop:
            break
        elif step < 0 and temp <= stop:
            break
        yield temp  #El yield reemplaza el return en el generador para este tipo de casos con float
        count += 1

def formula_y():
  x = i
  y = 15 - (5*x) + ((3*x)**2)
  return y
  print("%g"%y, end=", ")
```
